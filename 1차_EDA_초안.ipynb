# 5. 분포 분석

## 5.1. Binary 변수 불균형 파악

- 빈도가 높은 클래스에 집중하여 학습 -> 적은 클래스에 대한 예측은 제대로 학습하지 못함
- 적은 클래스에 대한 예측 성능은 낮아도, 다수 클래스의 예측만 잘하면 높은 정확도가 나옴.
- 변수 중요도, 상관 분석 등을 통해 예측에 미치는 영향을 확인하고, 제거 여부 결정

```python
import seaborn as sns
import matplotlib.pyplot as plt

# 고유 값이 2개인 변수들을 리스트로 수집
bi_col = [col for col in X.columns if X[col].nunique() == 2]

# 그래프 행과 열 개수 설정 (필요에 따라 조정)
n_cols = 3  # 한 행에 3개의 그래프씩
n_rows = (len(bi_col) + 1) // n_cols  # 필요한 행의 수 계산

# 서브플롯 생성 (figure 크기 조정 가능)
# fig: 전체 subplot을 담고 있는 figure 객체.
# axes: subplot들을 담고 있는 배열 또는 행렬 -> [[axes[0,0], axes[0,1]], [axes[1,0], axes[1,1]]] 이런 꼴
fig, axes = plt.subplots(n_rows, n_cols, figsize=(9, n_rows * 3))
axes = axes.flatten()  # [[axes[0], axes[1], axes[2], axes[3]] 이와 같이 변환

# 각 고유 값이 2개인 변수에 대해 그래프 그리기
for idx, col in enumerate(bi_col):
    sns.countplot(x=X[col], hue=X[col], palette='Set2', ax=axes[idx])  # 각 subplot에 그래프 그리기
    axes[idx].set_title(f'{col}')  # 각 그래프의 제목 설정

# 그래프 사이의 레이아웃 조정
plt.tight_layout()
plt.show()
```

## 5.2. Categorical 변수 분포 파악

```python
cat_col = [col for col in X.columns if X[col].nunique() > 2]
print(len(cat_col))
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

n_cols, n_rows = 5, 3

fig, axes = plt.subplots(n_rows, n_cols, figsize=(15, n_rows * 3))
axes = axes.flatten()

for idx, col in enumerate(cat_col):
    sns.countplot(x=X[col], hue=X[col], palette='Set2', ax=axes[idx])
    axes[idx].set_title(f'{col}')  # 각 그래프의 제목 설정

# 그래프 사이의 레이아웃 조정
plt.tight_layout()
plt.show()
```
